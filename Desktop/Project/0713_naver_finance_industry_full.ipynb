```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd
import requests
from io import StringIO
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfparser import PDFParser
import os
from naver_crawler import *
```

```python
driver = webdriver.Chrome('C:/Users/shic/Desktop/Project/chromedriver.exe')
driver.implicitly_wait(3)
start_date='2021-01-01'
end_date='2021-07-14'
pages = list(range(1,80,1)) # 2020년 1월 ~ 2021년 7월
print(pages)
```

```python
# 네이버 산업분석 리포트 크롤링

idx_list = [2,3,4,5,6,10,11,12,13,14,18,19,20,21,22,26,27,28,29,30,34,35,36,37,38,42,43,44,45,46]
df = pd.DataFrame(columns = ['date','kapital','title','pdf','text'])
end_page=728

for p in pages:
    url = "https://finance.naver.com/research/industry_list.nhn?&page={}".format(p)
    html = req.urlopen(url)
    html.encoding = 'cp949'
    doc = BeautifulSoup(html,"html.parser")
    
    srlists = doc.find_all('tr')
    
    for i in range(30):
        tr = srlists[idx_list[i]]
        # get data
        date = '20'+tr.find_all('td')[4].text.replace('.','')
        if idx_list[i] == len(srlists)-1:
            end_flag = True
            break
        kap = tr.find_all('td')[2].text
        pdf = tr.find_all('td')[3].find('a')['href'][-17:-4]
        link = tr.find('td', class_='file').find('a').attrs['href']
        fire_res = requests.get(link)

        with open('ind_report{}_{}.pdf'.format(p,i),'wb') as f:
            f.write(fire_res.content)
        os.chdir('C:\\Users\\shic\\Desktop\\Project\\ind_report')
        text = read_pdf_PDFMINER('ind_report{}_{}.pdf'.format(p,i))
        
        # print(text)
        href = 'https://finance.naver.com/research/'+tr.find_all('td')[1].find('a')['href']
        print(date, kap, pdf, href,link)
        
        # get data from second url
        html2 = urlopen(href)
        sour2 = BeautifulSoup(html2.read(), "html.parser")

        title = sour2.find_all("th")[0]
        a1 = str(title).find('span')
        a2 = len(str(title.find('span')))
        b1 = str(title).find(str(title.find('p')))
        title = str(title)[a1+a2:b1].replace('\n','')
        title = title.replace('\t','')
        
        record  = pd.Series([date, kap, title, pdf,text],index = ['date','kapital','title','pdf','text'])

        # print(record)
        df = df.append(record, ignore_index = True)

```
