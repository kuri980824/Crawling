```python
browser = webdriver.Chrome('C:/Users/shic/Desktop/Project/chromedriver.exe')
url="http://www.seibro.or.kr/websquare/control.jsp?w2xPath=/IPORTAL/user/company/BIP_CNTS01042V.xml&menuNo=286#"
browser.get(url)
```

```python
#브라우저에서 조회기간을 2020년, 조회기준을시가배당율 으로 선택한다.
search_btn=browser.find_element_by_id("image1")
search_btn.click()
result = browser.page_source
soup = BeautifulSoup(result, 'html.parser')
```

```python
df = pd.DataFrame(columns = ['순위','종목코드','종목명','주식종류','시장구분','주당배당금','시가배당율','액면가배당율','액면가','결산월'])
df
```

```python
for lk in range(2,100):
    lk_button = browser.find_element_by_id("cntsPaging01_next_btn")
    lk_button.click()
    for n in range(15):
        result = browser.page_source
        soup = BeautifulSoup(result, 'html.parser')
        rank = soup.find_all('td',{'id':'grid1_cell_{}_0'.format(n)})[0].text
        stk_code = soup.find_all('td',{'id':'grid1_cell_{}_1'.format(n)})[0].text
        stock = soup.find_all('td',{'id':'grid1_cell_{}_2'.format(n)})[0].text
        stk_type = soup.find_all('td',{'id':'grid1_cell_{}_3'.format(n)})[0].text
        mkt_type =soup.find_all('td',{'id':'grid1_cell_{}_4'.format(n)})[0].text
        dbd = soup.find_all('td',{'id':'grid1_cell_{}_5'.format(n)})[0].text
        mdp = soup.find_all('td',{'id':'grid1_cell_{}_6'.format(n)})[0].text
        fdp =soup.find_all('td',{'id':'grid1_cell_{}_7'.format(n)})[0].text
        fv = soup.find_all('td',{'id':'grid1_cell_{}_8'.format(n)})[0].text
        cal_mth = soup.find_all('td',{'id':'grid1_cell_{}_9'.format(n)})[0].text
    
        record  = pd.Series([rank,stk_code,stock,stk_type,mkt_type,dbd,mdp,fdp,fv,cal_mth],index = ['순위','종목코드','종목명','주식종류','시장구분','주당배당금','시가배당율','액면가배당율','액면가','결산월'])

        df = df.append(record, ignore_index = True)
```
