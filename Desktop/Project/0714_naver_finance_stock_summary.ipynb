```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd
import requests
from io import StringIO
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfparser import PDFParser
import os
from naver_crawler import *
```

```python
driver = webdriver.Chrome('C:/Users/shic/Desktop/Project/chromedriver.exe')
driver.implicitly_wait(3)
# start_date='2020-01-01' end_date='2021-07-14'
pages = list(range(1,80,1)) # 2020년 1월 ~ 2021년 7월
print(pages)
```

```python
idx_list = [2,3,4,5,6,10,11,12,13,14,18,19,20,21,22,26,27,28,29,30,34,35,36,37,38,42,43,44,45,46]
df = pd.DataFrame(columns = ['date', 'stock_name', 'target_price', 'invest_opinion', 'kap', 'title', 'href', 'summary'])
count=0
```

```python
df
```

```python
# 네이버 종목분석 리포트 크롤링

for p in pages:
    
    url = "https://finance.naver.com/research/company_list.nhn?&page={}".format(p)
    html = req.urlopen(url)
    html.encoding = 'cp949'
    doc = BeautifulSoup(html,"html.parser")

    srlists = doc.find_all('tr')
  
    for i in range(30):
        
        tr = srlists[idx_list[i]]

        
        date = '20'+tr.find_all('td')[4].text.replace('.','')
        stock_name = tr.find('a').text
        kap = tr.find_all('td')[2].text
        title = tr.find_all('td')[1].text
        href = 'https://finance.naver.com/research/'+tr.find_all('td')[1].find('a')['href']

        # get data from second url
        html2 = urlopen(href)
        sour2 = BeautifulSoup(html2.read(), "html.parser")
        summary = sour2.find("td",'view_cnt').find('div').text.replace('\n','')
        target_price = sour2.find('em',{'class':'money'}).text
        invest_opinion = sour2.find('em',{'class':'coment'}).text
        
        # print(date, kap, pdf, title, href, link, summary)
        record  = pd.Series([date, stock_name, target_price, invest_opinion, kap, title, href, summary],index = ['date', 'stock_name', 'target_price', 'invest_opinion', 'kap', 'title', 'href', 'summary'])
        print(record)
        
        df = df.append(record, ignore_index = True)
        
        count+=1
        if count%100==False:
            print(count)        
```
